### Import dataset

Imported clean dataset for visualization

```python
# Read dataset cleaned
dataset = pd.read_csv('Clean_Dataset.csv')
print(dataset)
```

```python
# dataset.head() prints/shows first 5 rows of dataset
dataset.head()
```

```python
dataset.drop('Unnamed: 0', axis=1, inplace=True)
dataset.head()
```

```python
# Segregate/Separate the data into Numeric and Catergorical ones
categorical, numeric = [], []
for ele in dataset.columns:
    if dataset[ele].dtype == 'object':
        categorical.append(ele)
    else:
        numeric.append(ele)
print("Categorical Data : ", categorical)
print("Numeric Data : ", numeric)
```

```python
dataset.drop(['condition_code', 'neo', 'pha','n_obs_used', 'data_arc', 'class'], axis = 1)
```

## Data Visualization 

Heatmap of Correlation with Annotation

```python
plt.figure(figsize=(14,12))
sns.heatmap(dataset.corr(), annot=True)
plt.show()
```

### Feature 'n' is having negative correlation with many other features 

Negative Correlation: variables change in opposite directions.

A negative linear relationship is observed when the values of variables have trends that occur in opposite directions (inverse relationship). That is, as the values of one variable increase the values of the other variable tend to decrease.

While exploring the data, if we want to discover possible correlations between the attributes of our data.

A scatter plot is probably a good way to visualize this: it allows you to identify a relationship between the two features that you have gained from the dimensionality reduction.

### Scatter Plot 
Scatter Plot can be used with several semantic groupings which can help to understand well in a graph. They can plot two-dimensional graphics that can be enhanced by mapping up to three additional variables while using the semantics of hue, size, and style parameters

##### A Scatter Plot of feature 'a' VS 'n'

```python
plt.scatter(dataset['a'], dataset['n']) # Customize the plot
plt.title("a VS n") # Customize the plot
plt.xlabel("a values") # Customize the plot
plt.ylabel("n values") # Customize the plot
plt.show() # Show the plot
```

```python
plt.scatter(dataset['n'], dataset['per_y']) # Customize the plot
plt.title("n VS per_y") # Customize the plot
plt.xlabel("n values") # Customize the plot
plt.ylabel("per_y values") # Customize the plot
plt.show() # Show the plot
```

A perfect Negative Correlation is visible

### Counter PLot 

A contour plot is a graphical method to visualize the 3-D surface by plotting constant Z slices called contours in a 2-D format. The contour plot is an alternative to a 3-D surface plot.

A Counter Plot of feature 'class'

```python
sns.countplot(x='class', data=dataset)
plt.title("Class Count PLot") # Customize the plot
plt.show() # Show the plot
```

### Box Plot 

Box Plot: It is a type of chart that depicts a group of numerical data through their quartiles. 

It is a simple way to visualize the shape of our data. It makes comparing characteristics of data between categories very easy.

A Box Plot of feature 'class' VS 'albedo'

```python
sns.catplot(x='class', y='Diameter', kind='box', data=dataset)
plt.title("class VS Diameter") # Customize the plot
plt.show() # Show the plot
```

### Kde(Kernel Density Estimate) plot 

KDE Plot described as Kernel Density Estimate is used for visualizing the Probability Density of a continuous variable.

It depicts the probability density at different values in a continuous variable. We can also plot a single graph for multiple samples which helps in more efficient data visualization.

A Kde Plot of feature 'Diameter' VS 'Density'

```python
sns.kdeplot(x='Diameter', data=dataset)
plt.title("Diameter VS Density") # Customize the plot
plt.show() # Show the plot
```

A Kde Plot of feature 'Diameter' VS 'Density' with hue as 'class'

```python
plt.figure(figsize=(10, 10)) # Customize the plot
plt.title('Diameter VS Density') # Customize the plot
sns.kdeplot(x='Diameter', hue='class', data=dataset)
plt.show() # Show the plot
```

### Histogram 

A histogram is an accurate graphical representation of the distribution of numerical data. 
It is an estimate of the probability distribution of a continuous variable.

A Distribution Histogram for feature 'a'

```python
plt.hist(dataset['a']) # Customize the plot
plt.ylabel("No. of Values") # Customize the plot
plt.xlabel("a values") # Customize the plot
plt.title("a Distribution Histogram") # Customize the plot
plt.show() # Show the plot
```

A Distribution Histogram for feature 'i'

```python
sns.histplot(x='i', data=dataset, bins=6, hue='class')
plt.title("i Distribution Histogram") # Customize the plot
plt.show() # Show the plot
```
